## Теорема Фишера для музыкальных жанров

В книге «Эгоистичный ген» Ричард Докинз высказал гипотезу о том, что эволюция идей в человеческой культуре происходит в целом идентично эволюции биологической, и предложил понятие _мема_ как культурного аналога гена -- наименьшей биологической единицы, способной к репликации. Идея Докинза является, в сущности, лишь метафорой, она открывает возможность для исследования культурных явлений методами биологии. В данной работе мы, хотя и не используем идею генов напрямую, проводим исследование в таком духе, а именно, пытаемся выяснить, выполняется ли теорема Фишера для «популяций» музыкальных жанров. 

Теорема Фишера о естественном отборе - один из фундаментальных результатов математической популяционной биологии, сформулированный Рональдом Фишером в 1930 году в книге «Генетическая теория естественного отбора». Формулируется эта теорема следующим образом: 

>_Интенсивность отбора (и, следовательно, скорость эволюции путем отбора) пропорциональна величине генетической дисперсии  по приспособленности эволюционирующей популяции, которая, в свою очередь, пропорциональна эффективному размеру популяции_ (Кунин Е. Логика случая. М, Центрполиграф, 2010. - стр. 24). 

Интуитивно, музыкальный жанр можно соотнести с популяцией, которая пытается приспособиться ко вкусу слушателей, и в таком случае количество релизов можно отождествить с эффективным размером популяции, а количество прослушиваний - с приспособленностью. Мы собираемся проверить, насколько «эволюция» музыкальных жанров подчиняется теореме Фишера, иначе говоря, какую роль в ней играет «естественный отбор» слушателями. Далее мы приводим две точные формулировки теоремы Фишера, принадлежащие Прайсу и Баезу, и обсуждаем в их контексте операционализацию наших музыкальных величин.  


#### Теорема Фишера по Прайсу

Джордж Прайс в статье «Fisher's 'fundamental theorem' made clear» (https://doi.org/10.1111/j.1469-1809.1972.tb00764.x) восстанавливает оригинальную формулировку Фишера в следующем виде. 
Пусть $M(S,E):\mathbb R^2\to \mathbb R$ обозначает среднюю приспособленность популяции, которая зависит от двух аргументов: естественного отбора $S(t)$ (selection) и влияния среды $E(t)$ (environment). Тогда $\dot M = M'_S \dot S + M'_E \dot E$. С другой стороны, представим среднюю приспособленность в виде $M(t) = bQ$, где $Q(t) = (q_1(t),...,q_n(t))$ - вектор частот аллелей генов в популяции, $b = (b_1,...,b_n)$ - коэффициенты регрессии приспособленности на частоту аллелелей. Отсюда получем, что $\dot M = \dot b Q + b\dot Q$.  
По Фишеру, величина $b\dot Q$ выражает ту часть изменения $M$, которая вызвана естественным отбором, или же линейным изменением частот генов, а величина $\dot bQ$ отвечает за изменения среды. Отсюда мы получаем два равенства
$$ \begin{matrix}
M'_S \dot S = b\dot Q, \\ M'_E \dot E = \dot bQ.
\end{matrix}
$$
Эмпирически мы можем вычислить $M(t)$ и $Q(t)$, а далее подобрать коэффициенты $b(t)$ и вычислить производные всех величин, взяв конечные разности, поэтому нетривиальная гипотеза, которую мы будем проверять, такова: «приспособленность» музыки управляется «естественным отбором» по её параметрам. Успешность этой гипотезы можно измеряется мерой
$$ \frac{M'_S \dot S}{M'_S \dot S + M'_E \dot E} \quad(1)$$
Если эта величина близка к 1, то эволюция жанра определяется естественным отбором, если она близка к нулю - то случайными факторами. 

Величина $M'_S \dot S = b\dot Q =: W(t)$ также называется аддитивной генетической дисперсией по приспособленности. Соответственно, вторая часть гипотезы состоит в том, что $W(t)$ пропорциональна эфффективному размеру популяции. 

В гипотезе Фишера гены имеют дискретное значение (есть конечное число аллелей в данном локусе). В качестве геномных локусов мы рассматриваем характеристики музыки, доступные в датасете `spotify`, а именно explicit, danceability, energy, loudness, speechiness, acousticness, instrumentalness, liveness, valence, tempo. Все они (кроме explicit и liveness) представлены в датасете вещественными числами в промежутке $[0,1]$. Мы дискретизируем их простейшим образом, а именно превращает в тип bool по следующему правилу: $[0,0.5)\to False,[0.5,1]\to True$.           



#### Теорема Фишера по Баезу

Другую формулировку теоремы Фишера, несколько отстоящую от оригинальной, предложил Джон Баез (https://math.ucr.edu/home/baez/fisher/fisher_web.pdf). Пусть $P_i(t):\mathbb R\to \mathbb R, i=1,...,n$ - численности популяций. Не накладывая пока никаких ограничений, назовем функции $f_i(P_1,...,P_n)$ функциями приспособленности. Уравнение Лотки-Вольтерра описывает динамику численности популяций с заданными приспособленностями: $\dot P_i = f_i(P_1,...,P_n)P_i$. 
Пусть $p_i = P_i / \sum_j P_j$ - вероятность для организма принадлежать $i$-той популяции. Тогда средняя приспособленность равна $\langle f \rangle = \sum_j f_jp_j$, а дисперсия приспособленности $Var(f) = \sum_j p_j(f_j - \langle f \rangle)^2$. В данных обозначениях теорема Фишера утверждает, что 
$$\frac{d\langle f \rangle}{dt} = Var(f),\quad (1')$$
однако это верно, если и только если все $f_i = const$. 
Баез доказывает следующую теорему: для вектора $p = (p_1,...,p_n)$
имеет место 
$$\|\dot p\|^2 = Var(f).\quad (2')$$

В таком сеттинге мы можем изучать ситуацию в целом. Положим $P_i(t) = $ количество треков жанра $i$, вышедших в момент времени $t$, приспособленность жанра $f_i$ равна количеству прослушиваний в момент $t$. В таком виде мы можем проверить обе формулы $(1'),(2')$. Гипотеза нетривиальна, поскольку мы операционально определяем и приспособленность, и численности популяций.

```python
# ── 2. Загрузка Spotify ──────────────────────────────────────────
import pandas as pd
from datasets import load_dataset

spotify = pd.read_csv("data/spotify/dataset.csv", index_col=0)


maxmin_values = spotify.agg(['min','max'])


#print(max_values)
print(maxmin_values)

```

```python
# Переименуем для удобства
spotify = spotify.rename(columns={
    "track_name": "track_name",
    "artists":    "artist_name",
    "track_genre": "genre",
})

# Оставим нужные фичи
spotify_cols = [
    "track_name", "artist_name", "genre",
    "explicit", "danceability", "energy",
    "loudness", "speechiness", "acousticness",
    "instrumentalness", "liveness", "valence", "tempo"
]
spotify = spotify[spotify_cols].drop_duplicates(subset=["track_name", "artist_name"])

print(f"Spotify: {len(spotify):,} треков")
print(spotify.head())
```

```python
# Бинаризуем численные признаки: 

numeric = spotify.select_dtypes(include='number').columns
threshold = (spotify[numeric].max() - spotify[numeric].min()) / 2
spotify[numeric] = spotify[numeric] > threshold

print(spotify.head())
```

```python
# ── 1. Загрузка Last.fm ──────────────────────────────────────────
lastfm = pd.read_csv(
    "./data/lastfm/lastfm-dataset-1K/userid-timestamp-artid-artname-traid-traname.tsv",
    sep="\t",
    names=["user_id", "timestamp", "artist_mbid", "artist_name", "track_mbid", "track_name"],
    header=None,
    on_bad_lines="skip",
)

# Приводим timestamp к datetime
lastfm["timestamp"] = pd.to_datetime(lastfm["timestamp"], utc=True, errors="coerce")

print(f"Last.fm: {len(lastfm):,} строк")
print(lastfm.head())


# ── 3. Нормализация строк перед джойном ─────────────────────────
def normalize(s: pd.Series) -> pd.Series:
    return (
        s.str.lower()
         .str.strip()
         .str.replace(r"[^\w\s]", "", regex=True)  # убираем пунктуацию
         .str.replace(r"\s+", " ", regex=True)
    )

lastfm["_track"]  = normalize(lastfm["track_name"])
lastfm["_artist"] = normalize(lastfm["artist_name"])

spotify["_track"]  = normalize(spotify["track_name"])
spotify["_artist"] = normalize(spotify["artist_name"])

```

```python
# ── 4. Джойн ────────────────────────────────────────────────────
merged = lastfm.merge(
    spotify,
    on=["_track", "_artist"],
    how="inner",
    suffixes=("_lfm", "_sp"),
)

# Убираем служебные колонки
merged = merged.drop(columns=["_track", "_artist"])

print(f"\nРезультат джойна: {len(merged):,} строк")
print(f"Покрытие: {len(merged)/len(lastfm)*100:.1f}% от Last.fm событий")
print(merged.head())


# ── 5. Сохраняем ────────────────────────────────────────────────
merged.to_parquet("data/merged_listening_history.parquet", index=False)
print("\nСохранено в data/merged_listening_history.parquet")
```

## Проверка гипотезы (1)

Перейдём к тестированию гипотез. Первую гипотезу мы проверим на наиболее популярных жанрах, разбив промежуток 2005-2009 на меясцы. Основные величины определяются следующим образом:

- Положим $dt =1$;
- $M(t) = \frac1N(t) \sum\limits^{N(t)}_{j=1} w_j(t)$, где $N(t)$ - количество треков, послушанных в период $t$, $w_j(t)$ - количество прослушиваний j-того трека за период $t$ (получаем средунюю приспособленность по жанру);
- $\dot M(t) = M(t+1) - M(t)$ - скорость изменения средней приспособленности;
- $Q_i(t) = \frac1N(t)\sum\limits^{N(t)}_{j=1} x_{ij}$, где $x_{ij} =1$, если у $j$-того трека $i$-тый признак True, 0 если False (заметим, что $Q$ зависит от $t$ только потому, что в разные моменты могут быть послушаны разные треки);  
- $b_i(t)$ = линейная регрессия прослушиваний от признаков; 
- $M'_S\dot S = b(t)Q(t+1) - b(t)Q(t)$ - влияние отбора;
- $M'_E\dot E = b(t+1)Q(t+1) - b(t)Q(t+1)$ - влияние среды;
- $ \dfrac{b(t)Q(t+1) - b(t)Q(t+1)}{ b(t+1)Q(t+1) - b(t)Q(t) } $ - мера влияния отбора;



```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# 1. Список бинарных признаков
features = ['danceability', 'energy', 'explicit', 'acousticness', 'loudness', 'speechiness', 'instrumentalness',
            'liveness','valence','tempo']

genre = 'mandopop'

track_col, artist_col = 'track_name_sp', 'artist_name_sp'

results = []
months = sorted(merged['month'].unique())

for i in range(len(months) - 1):
    t0, t1 = months[i], months[i+1]
    
    # Фильтруем данные по месяцам
    df_t0 = merged[(merged['month'] == t0) & (merged['genre'] == genre)]
    df_t1 = merged[(merged['month'] == t1) & (merged['genre'] == genre)]
    
    def get_population_stats(df):
        if df.empty:
            return pd.Series(0, index=features), 0, pd.Series(0, index=features), 0

    # 1. Группируем данные: для признаков берем первое значение, 
    # а количество строк (прослушиваний) считаем отдельно
        tracks = df.groupby([track_col, artist_col]).agg({f: 'first' for f in features})
        tracks['listens'] = df.groupby([track_col, artist_col]).size()
    
    # 2. Q_t: среднее значение признаков среди АКТИВНЫХ треков
        q_vec = tracks[features].mean()
    
    # 3. M_t: среднее количество прослушиваний на один активный трек
        m_val = tracks['listens'].mean()
    
    # 4. b_t: ценность признаков через регрессию
        if len(tracks) > len(features):
            model = LinearRegression()
            model.fit(tracks[features], tracks['listens'])
            b_vec = pd.Series(model.coef_, index=features)
            intercept = model.intercept_
        else:
            b_vec = pd.Series(0, index=features)
            intercept = 0
         
        return q_vec, m_val, b_vec, intercept

    q_t0, m_t0, b_t0, c_t0 = get_population_stats(df_t0)
    q_t1, m_t1, b_t1, c_t1 = get_population_stats(df_t1)
    
    # Считаем компоненты Прайса
    # 1. Отбор (изменение состава активных треков)
    delta_ns = (b_t0 * (q_t1 - q_t0)).sum()
    
    # 2. Среда (изменение вкусов слушателей и весов признаков)
    delta_ec = ((b_t1 - b_t0) * q_t1).sum() + (c_t1 - c_t0)
    
    results.append({
        'month': t1,
        'delta_ns': delta_ns,
        'delta_ec': delta_ec,
        'm_val': m_t1
    })

price_df = pd.DataFrame(results)

import matplotlib.pyplot as plt

# Рассчитываем долю влияния отбора (Ratio)
# Используем абсолютные значения, чтобы понять вклад в общую вариативность
price_df['selection_ratio'] = price_df['delta_ns'].abs() / (price_df['delta_ns'].abs() + price_df['delta_ec'].abs())

# Сгладим график (скользящее среднее), так как месячные данные могут быть очень шумными
price_df['selection_ratio_smooth'] = price_df['selection_ratio'].rolling(window=3, center=True).mean()

plt.figure(figsize=(12, 6))

# Рисуем основную линию и сглаженную
plt.plot(price_df['month'].astype(str), price_df['selection_ratio'], 
         alpha=0.3, color='gray', label='Месячное значение')
plt.plot(price_df['month'].astype(str), price_df['selection_ratio_smooth'], 
         color='blue', linewidth=2, label='Сглаженное (3 мес.)')

# Добавляем эталонную линию 0.5
plt.axhline(0.5, color='red', linestyle='--', label='Порог 50% (Граница доминирования)')

plt.title('Доля влияния "Естественного отбора" в динамике Mandopop')
plt.ylabel('Вклад отбора в общее изменение (Ratio)')
plt.xlabel('Месяц')
plt.ylim(0, 1) # Доля всегда от 0 до 1
plt.legend()
plt.xticks(rotation=45)
plt.grid(True, alpha=0.2)
plt.tight_layout()
plt.show()

# Печатаем среднее значение для вывода
print(f"Средняя доля влияния отбора: {price_df['selection_ratio'].mean():.2%}")
```
